```python
htmlcontent = r.content
soup = BeautifulSoup(htmlcontent, 'html.parser')
print(soup)
```

```python
anchor_same_class  =  soup.find('td') # to find specific class inside anchor tag
print(anchor_same_class)
```

```python
price = soup.select("#productTitle")
print(price)
```

```python
links = soup.find_all("a", attrs={'class':'a-link-normal s-no-outline'})
print(links)
```

```python
links_list = []
for link in links:
    links_list.append(link.get('href'))
    print(link.get('href'))
    print()
    
        
#print(links_list)
```

```python
def get_title(soup):
    try:
        title_string = soup.find("span", attrs={"class":'a-size-large product-title-word-break'}).string.strip()
        #title_value = title.string
        #title_string = title_value.strip()
    except AttributeError:
        title_string = ""
    #print(title_string)
        
    return title_string

def get_price(soup):
    try:
        price = soup.find("span", attrs={"id":'priceblock_ourprice'})
        price_value = price.string
        price_string = price_value.strip()
    except AttributeError:
        try:
            price = soup.find("span", attrs={'id':'priceblock_dealprice'}).string.strip()
            price_value = price.string
            price_string = price_value.strip()
        except:
            price_string = ''
    return price_string

def get_rating(soup):
    try:
        rating = soup.find("i", attrs={'class':'a-icon a-icon-star a-star-4-5'}).string.strip()
    except:
        rating = ''
    return rating

def get_description(soup):
    try:
        desc = soup.find_all('div', attrs={'id':'productDescription'})
    except:
        desc = ''
    return desc


        
            
for link in links_list:
    new_webpage  = requests.get("https://www.amazon.in" + link, headers=headers)
    new_soup = BeautifulSoup(new_webpage.content, "lxml")
    print("----tilte= " ,get_title(new_soup))
    print("price  = ",get_price(new_soup) )
    print("rating  = ",get_rating(new_soup) )
    print("Description = ", get_description(new_soup))
    print()
   
```
